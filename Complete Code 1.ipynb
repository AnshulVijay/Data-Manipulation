```python
from __future__ import print_function
import string
import re
import codecs
from collections import Counter
```

```python
# to return a sorted list of words extracted from the book's text_file that also exist in the Sowpods list
def create_wordlist(text_file):
    file = codecs.open(text_file,"r",encoding='utf8')
    text = file.read()
    file.close()
# to convert all characters to lowercase
    stripped = text.lower()
    
# replace hyphens with a single space
    stripped = stripped.replace("-"," ")
    
# strip off all contractions and possessives from words
    stripped = re.sub('\'[a-z]+','',stripped)
    
# remove all punctuation
    stripped1 = stripped.maketrans("","",string.punctuation)
    stripped = stripped.translate(stripped1)
    words = stripped.split()
    
    
# remove all whitespace characters and numbers
    stripped1 = stripped.maketrans("","",string.whitespace)
    stripped = stripped.translate(stripped1)

# to find all the unique words in the text
    unique_words = []
    for i in words:
        if not i in unique_words:
            unique_words.append(i)

    file1 = open("sowpods.txt","rt")
    text1 = file1.read()
    file1.close()
    final = []
    stripped2 = text1.lower()
    words1 = stripped2.split()
    word3 = []
    
# to find all the words common in text file and sowpods file
    for i in range(len(unique_words)):
        if unique_words[i] in words1:
            word3.append(unique_words[i])
    word3.sort()
    return(word3)
```

```python
# to find that is the word can be matched in letters given
def word_match(word, letters):
    word2, word1 = Counter(word), Counter(letters)
# checking each letter of word by matching it in letters
    return(all(word2[k] <= word1.get(k,0) for k in word2))
```

```python
# to find the word score of the given word
def word_score(word):
# creating two list ,one for alphabets and its values
    list2 = [1]*26
    list1 = ["a","e","i","o","n","r","t","l","s","u","d","g","b","c","m","p","f","h","v","w","y","k","j","x","q","z"]
    list2 = list1.copy()
    list2[:10] = [1]*10
    list2[10:12] = [2]*2
    list2[12:16] = [3]*4
    list2[16:21] = [4]*5
    list2[21:22] = [5]*1
    list2[22:24] = [8]*2
    list2[24:26] = [10]*2
    count = 0
    for i in range(len(word)):
# getting the value after matching the letter
        count += list2[list1.index(word[i])]
    return(count)
```

```python
# to find the words from given words_list which can be formed from the letters given
def find_words(words_list, letters):
    final_words = {}
# checking word list one by one using pre define word_match function
    for i in range(len(words_list)):
        flag1 = word_match(words_list[i],letters)
        if flag1==True:
        # updating the word matched in a new list
            final_words.update({words_list[i]:word_score(words_list[i])})
    return(final_words)
```

```python
# to find the unique only words in the given text file
def a1(text_file):
    file = open(text_file,"rt")
    text = file.read()
    file.close()
    # performing all the steps to find the real unique letters in the text
    strippedtext = text.lower()
    strippedtext = re.sub('\'[0-9]+','',strippedtext)
    stripped1 = strippedtext.maketrans("","",string.punctuation)
    strippedtext = strippedtext.translate(stripped1)
    words = strippedtext.split()
    unique_words = []
    for i in words:
        if not i in unique_words:
            unique_words.append(i)
    return(unique_words)
```

```python
# to find the word from the text_file that can be formed from the given letters
def a2(text_file, letters):
    words_text = a1(text_file)
    final = {}
    maxword = {}
    # matching the words that can be formed from given letters present in th text_file
    for i in range(len(words_text)):
        flag1 = word_match(words_text[i],letters) 
        if flag1==True:
            final.update({words_text[i]:word_score(words_text[i])})
    # updating the words that can be formed in a new dict with word score
    for i in range(len(final)):
        maxword.update({max(final, key = final.get):word_score(max(final, key = final.get))})
    return(maxword)
```

```python
# to find the counts how many words in a book contain certain letters
def a3(book_list, letters):
    textbook = () 
    for k in range(len(book_list)):
        count = 0
        totalwords = []
        totalwords = a1(book_list[k][1])
        # searching the whole words in the text_file using the word_match 
        for j in range(len(letters)):
            for i in range(len(totalwords)):
                flag1 = word_match(letters[j],totalwords[i])
            # counting the total no of words with given letters
                if flag1==True:
                    count = count + 1
        # deleting repeated count
        flag1 = word_match(letters,totalwords[i])
        if flag1==True:
            count = count - 1   
        textbook = textbook + (book_list[k][0],count)
    return(textbook)
```
