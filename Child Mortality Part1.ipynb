```python
# importing required modules for the task
import pandas as pd
import operator
import matplotlib.pyplot as plt
```

```python
# reading the Mortality rate pdf
new=pd.read_csv(r'A8-WHOSIS_MDG_000003.csv')
Country = new.iloc[:,0]
Year = new.iloc[:,1]
Infant_mortality = new.iloc[:,2]
Neonatal_mortality = new.iloc[:,3]
UnderFive_mortality = new.iloc[:,4]
```

```python
new.head()
```

```python
# seperating the mortality rates from the confidence interval

# for Infant mortality Rates
Final_Infant_mortality = []
for i in range(len(Infant_mortality)):
    u=0
    for j in range(len(Infant_mortality[i])):
        u=u+1
        if(str(Infant_mortality[i][j])!='['):
            continue
        else:
            Final_Infant_mortality.append(float(Infant_mortality[i][:j-1]))
            
# for Neonantal Mortality Rates
Final_Neonatal_mortality = []
for i in range(len(Neonatal_mortality)):
    u=0
    for j in range(len(Neonatal_mortality[i])):
        u=u+1
        if(str(Neonatal_mortality[i][j])!='['):
            continue
        else:
            Final_Neonatal_mortality.append(float(Neonatal_mortality[i][:j-1]))
           
# for Under Five Mortality Rates
Final_UnderFive_mortality = []
for i in range(len(UnderFive_mortality)):
    u=0
    for j in range(len(UnderFive_mortality[i])):
        u=u+1
        if(str(UnderFive_mortality[i][j])!='['):
            continue
        else:
            Final_UnderFive_mortality.append(float(UnderFive_mortality[i][:j-1]))
```

```python
import random

# getting Three random Countries from the Dataset to perform the comparisions
All_Country = new['Country'].unique()
All_Country_Random = []
for i in range(3):
    All_Country_Random.append(random.choice(All_Country))
All_Country_Random
```

```python
# Getting the Mortality Rates data for year 2016 for all Countries
Year_Data = []
for i in range(len(Year)):
    if(str(Year[i])=='2016'):
        dirc=[Final_Infant_mortality[i],Final_Neonatal_mortality[i],Final_UnderFive_mortality[i]]
        Year_Data.append(dirc)
```

```python
# Graph 1
# graph comparing five countries mortality rates
x = []
for i in range(5):
    x.append(All_Country[i])
y1 = []
y2 = []
y3 = []
for i in range(5):
    y1.append(Year_Data[i][0])
for i in range(5):
    y2.append(Year_Data[i][1])
for i in range(5):
    y3.append(Year_Data[i][2])

plt.plot(x,y1,label='Infant')
plt.plot(x,y2,label='Neonatal')
plt.plot(x,y3,label='UnderFive')
plt.legend()
```

```python
# Graph 2
# graph comparing the Infant Mortality Rates for three Random Countries for year 1990 to 2016
x = new['Year'].unique()
y1 = []
y2 = []
y3 = []
for i in range(len(Country)):
    if(str(Country[i])==All_Country_Random[0]):
        y1.append(Final_Infant_mortality[i])
for i in range(len(Country)):
    if(str(Country[i])==All_Country_Random[1]):
        y2.append(Final_Infant_mortality[i])
for i in range(len(Country)):
    if(str(Country[i])==All_Country_Random[2]):
        y3.append(Final_Infant_mortality[i])
        
plt.plot(x,y1, label=All_Country_Random[0])
plt.plot(x,y2, label=All_Country_Random[1])
plt.plot(x,y3, label=All_Country_Random[2])
plt.legend()
```

```python
from matplotlib.pyplot import figure

# Graph to compare the Mortality Rates of all Countries
x = []
for i in range(int((len(All_Country))/8)):
    x.append(All_Country[i])
#for i in range 
y1 = []
y2 = []
y3 = []

# Infant Rate Mortality
for i in range(int((len(Year_Data))/8)):
    y1.append(Year_Data[i][0])
figure(num=None, figsize=(30,20))
plt.bar(x,y1)
plt.xticks(rotation='45')
plt.show()

# Neonantal Rate Comparision
for i in range(int((len(Year_Data))/8)):
    y2.append(Year_Data[i][1])
figure(num=None, figsize=(30,20))
plt.bar(x,y2)
plt.xticks(rotation='45')
plt.show()

# Under Five Mortality Rate Comparision
for i in range(int((len(Year_Data))/8)):
    y3.append(Year_Data[i][2])
figure(num=None, figsize=(20,15))
plt.bar(x,y3)
plt.xticks(rotation='45')
plt.show()
```

```python
# Infant, Neonantal, Underfive mortality comparison for Random 1 Country
Comparision_Year = []
y1 = []
y2 = []
y3 = []
Country_Data = []
for i in range(len(Country)):
    if(str(Country[i])==All_Country_Random[0]):
        Comparision_Year.append(Year[i])
        dirc=[Final_Infant_mortality[i],Final_Neonatal_mortality[i],Final_UnderFive_mortality[i]]
        Country_Data.append(dirc)
for i in range(len(Country_Data)):
    y1.append(Country_Data[i][0])


for i in range(len(Country_Data)):
    y2.append(Country_Data[i][1])
    

for i in range(len(Country_Data)):
    y3.append(Country_Data[i][2])

plt.plot(Comparision_Year,y1,label='Infant')
plt.plot(Comparision_Year,y2,label='Neonatal')
plt.plot(Comparision_Year,y3,label='UnderFive')
plt.legend()
```

Infant Mortality Rates is mostly higher when compared to other Mortality Rates, as Infant Mortality Rate is the probability of dying between birth and exactly 1 year of age whereas Neonantal Mortality Rate is the probability of dying during first 28 days of life.

```python
# graph to compare the improvement in mortality rates over the 1990/2016 period for Random 1 Country

ratio = []
for i in range(len(Comparision_Year)):
    ratio.append(Country_Data[(len(Comparision_Year)-i-2)][1]/Country_Data[(len(Comparision_Year)-i-1)][1])

print(ratio)   
plt.bar(Comparision_Year,ratio)
#len(Comparision_Year)
#Country_Data
#Country_Data[(len(Comparision_Year))]
```
